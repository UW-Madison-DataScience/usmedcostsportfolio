#  Step 1: Importing libraries!

- In this first step, we will apply our expanded knowledge to import the following libraries: Seaborn, matplotlib and pandas, including the math library from before!

```python
import matplotlib.pyplot as plt #To create, show graphs!
import seaborn as sns # Used in conjunction with matplotlib to create visualizations!
import pandas as pd #Using this to read data and compute statistical info such as mean, median, range, etc.
import math # For basic mathematical functions, such as rounding a value to a few decimal places

insurance_data = pd.read_csv("insurancedata.csv")
insurance_data.head() # Displays the first five rows of data in insurancedata.csv file!
```

# Step 2: Explaining each variable

Let's talk about what each variable is below:
- Age: An integer number
- sex: A person's gender (male or female)
- bmi: Stands for Body Mass Index, indicates whether an individual is overweight, underweight or normal with this value
- children: Amount of children that an individual has(an integer)
- smoker: Whether an individual smokes or not 
- region: The section that an indivdual resides in the United States (southwest, southeast, northwest, northeast)
- charges: This is the amount of insurance costs that an individual typically pays

# Step 3: The Relationship between Region and Gender and Insurance Costs

To understand and make conclusions on this relationship between two categorical variables(region and gender) and a quantitative variable, we need to make a bar plot which will accurately shows key differences below:

```python
costs = sns.boxplot(data=insurance_data, x='charges', y='region', hue='sex', palette='winter')
costs.set_xlabel('Insurance Costs (in thousands)')
costs.set_ylabel('Regions')
costs.set_title('Insurance Costs between Regions & Gender')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.) # This is to make sure the legend is outside of the boxplot

```

# Conclusions and Findings

From the boxplot above, it definitely seems that the highest insurance costs from both genders are in the southeast region of the country.
Here are additional findings that I was able to summarize below:
- Overall, women from all regions tend to pay lower insurance costs than their male counterparts
- The highest insurance cost from this data also comes from the southeast, roughly around 61000-62000 dollars
- There are lots of outliers from all regions that stems from 20000 to just above 60000 in insurance costs
- The lowest insurance costs seems to stem around the range between 1500-2500 dollars




# Step 4: Relationship between Gender, BMI and Insurance Costs

- Since this time we are dealing with two quantitative variables and only one categorical variable(gender), we will have 
to create a different type of plot: a scatter plot!

```python
sns.scatterplot(data=insurance_data, x='bmi', y='charges', hue='sex', palette='spring')
plt.xlabel("Body Mass Index (BMI)")
plt.ylabel("Insurance Costs (in thousands)")
plt.title("Insurance Costs between Gender and BMI")
```

# Conclusions and Findings

- It's hard to tell but there is a weak positive correlation for the data above. Also, it can be inconclusive to determine whether having a higher BMI score leads to higher insurance costs since there are values with high BMIs(40-50s) that have very low and high costs.
Findings
- The majority of the individuals tend to have a BMI between 20 and 40 and have relative insurance costs hovering around 1000 and 10000 mark
- The highest insurance costs is from a female (over 60,000) with a BMI of roughly around 46-47, which would classify as obese



# Step 5: Relationship between Age and BMI

To figure this simple relationship between Age and BMI, we will create a relational plot(similar to a scatter plot) below:

```python
sns.set_theme(style='white')
age_bmi = sns.relplot(data=insurance_data, x='age', y='bmi')
plt.title("Age vs BMI")
plt.xlabel("Age")
plt.ylabel("Body Mass Index (BMI)")
```

# Conclusions & Findings

From the data, it seems that the distribution is not linear due to the points being scattered in no particular direction.
Due to that, it is difficult to determine whether age influences higher or lower bmi values.


# Step 6: Relationship between Insurance Costs & Smoker Status

To figure out the relationship between insurance costs & smoker status, we will create a bar plot that will display
the differences in costs for non-smokers and smokers:

```python
sns.barplot(data=insurance_data, x='smoker', y='charges', palette='spring')
plt.title("Insurance Costs vs. Smoker Status")
plt.xlabel("Smoker Status")
plt.ylabel("Insurance Costs (in thousands)")
```

# Conclusions & Findings

From the barplot, it definitely seems like individuals who smoke tend to have higher insurance costs than people who do not smoke.

Here is an additional finding:
- The highest insurance cost for people who do not smoke is roughly around 10,000 dollars while the highest cost for smokers 
is around $34,000

# Step 7: Finding an Association between children and insurance costs

To see if the amount of children a person has influences insurance costs, we need to create a box plot below:

```python
sns.boxplot(data=insurance_data, x='children', y='charges', palette='summer')
plt.xlabel("Amount of Children")
plt.ylabel("Insurance Costs (in thousands)")
plt.title("Insurance Costs vs. Number of Children")
```

# Conclusions and Findings

It definitely seems like having more children doesn't necessarily affect insurance costs, in fact it seems like people with more children tend to pay lesser amounts.
Additional Findings:
- Having no children nets both the highest (60,000+) and lowest (~$1500) amount of insurance costs!
- The average cost in each section seems to be slightly similar(with no children barely eeking out, most likely due to the vast 
amount of outliers!)
- Interestingly enough, since the boxplot that was made for people that have five children, the sample size is way smaller
than the others which may make the data seem inconclusive and in need of a larger subset



# Lastly: Additional Statistics from the Data

Below, I'll display some statistics such as median, max, min, mean, etc for you to see. Enjoy!

```python
# Getting the average, maximum, minimum values for insurance costs in the United States

avg_insurance = round(insurance_data.charges.mean(), 2) # Using the round function from the math library to cut down to 2 decimal places
print("The average insurance cost from this data is $" + str(avg_insurance))

max_insurance = round(insurance_data.charges.max(), 2)
print("The maximum insurance cost from this data is $" + str(max_insurance))

min_insurance = round(insurance_data.charges.min(), 2)
print("The minimum insurance cost from this data is $" + str(min_insurance))
```

```python
# Getting the average, maximum, minimum for BMIs in the United States

avg_bmi = round(insurance_data.bmi.mean(), 1)
print("The average bmi value from this data is: " + str(avg_bmi))

max_bmi = round(insurance_data.bmi.max(), 1)
print("The maximum bmi value from this data is: " + str(max_bmi))

min_bmi = round(insurance_data.bmi.min(), 1)
print("The minimum bmi value from this data is: " + str(min_bmi))
```
